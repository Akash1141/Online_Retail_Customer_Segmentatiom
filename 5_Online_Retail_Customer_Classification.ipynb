#### Chart - 13 - Radar Chart: Average Quantity and UnitPrice by Country

````python
# Chart - 13 visualization code
avg_quantity = df1.groupby('Country')['Quantity'].mean()
avg_unitprice = df1.groupby('Country')['UnitPrice'].mean()
categories = avg_quantity.index
N = len(categories)
values_quantity = avg_quantity.values.tolist()
values_unitprice = avg_unitprice.values.tolist()

angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
values_quantity += values_quantity[:1]
values_unitprice += values_unitprice[:1]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': 'polar'})
ax.plot(angles, values_quantity, linewidth=1, linestyle='solid', label='Quantity')
ax.plot(angles, values_unitprice, linewidth=1, linestyle='solid', label='Unit Price')
ax.fill(angles, values_quantity, alpha=0.25)
ax.fill(angles, values_unitprice, alpha=0.25)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
ax.legend(loc='upper right')
plt.title('Average Quantity and Unit Price by Country')
plt.show()

````

````python
# ML Model - 1 Implementation

# Predict on the model

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt

# Select the columns for clustering
columns = ['PC1', 'PC2' ]

# Create the K-means model
kmeans = KMeans(random_state=42)

# Define the parameter grid for hyperparameter tuning
param_grid = {'n_clusters': [2, 3, 4, 5]}

# Perform hyperparameter tuning using GridSearchCV
grid_search = GridSearchCV(kmeans, param_grid, cv=5)
grid_search.fit(df1[columns])

# Get the best model and its parameters
best_kmeans = grid_search.best_estimator_
best_params = grid_search.best_params_

# Fit the best model on the data
best_kmeans.fit(df1[columns])

# Predict the cluster labels
labels = best_kmeans.predict(df1[columns])

# Compute the silhouette score
#silhouette_avg = silhouette_score(df1[columns], labels)


````

### 1. Which Evaluation metrics did you consider for a positive business impact and why?

For K-means clustering, the evaluation metric commonly used is the silhouette score. The silhouette score measures how well each data point fits within its assigned cluster compared to other clusters. A higher silhouette score indicates that the data points are well-clustered and separated, which can lead to positive business impact. The silhouette score ranges from -1 to 1, with higher scores indicating better clustering.

For hierarchical clustering, the cophenetic correlation coefficient is often used as an evaluation metric. The cophenetic correlation coefficient measures the correlation between the pairwise distances of the original points and the pairwise distances of the clustered points. A higher cophenetic correlation coefficient suggests that the hierarchical clustering effectively preserved the pairwise distances of the data points, indicating better clustering quality.
